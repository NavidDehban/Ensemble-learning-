```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['cancer.csv']))
```

```python
df
```

```python
import numpy as np
features = df.columns[1:10]
f5 = df[features[5]]
avg = 0
cnt = 0
for i in range(len(f5)):
  if f5[i] != '?':
    avg += int(f5[i])
    cnt += 1
avg = avg/cnt

for i in range(len(f5)):
  if f5[i] == '?':
    f5[i] = avg

df[features[5]] = f5
```

```python
y = np.array(df['Class'])
x = np.array(df[features])
x = x.tolist()
y = y.tolist()
print(x)
print(y)
for i in range(len(x)):
  x[i][5] = int(x[i][5])
```

```python
print(x)
print(y)
print(len(x))
print(len(y))
```

```python
import numpy as np
from sklearn.ensemble import  VotingClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kf = KFold(n_splits=10)

clf1 = LogisticRegression()
clf2 = tree.DecisionTreeClassifier()
clf3 = svm.SVC()

eclf = VotingClassifier(estimators=[('lr', clf1), ('dt', clf2), ('svm', clf3)])
s = cross_val_score(eclf, x, y, scoring = 'accuracy', cv = kf)
print(s)
```

```python
print('accuracy:',sum(s)/len(s))
```
